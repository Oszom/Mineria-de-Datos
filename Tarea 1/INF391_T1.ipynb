```python
# Librerías útiles
import matplotlib.pyplot as plt
from numpy import genfromtxt
from sklearn.cluster import KMeans
```

```python
# Cargar los datos
d1 = genfromtxt('T1-1.csv', delimiter=',')
d2 = genfromtxt('T1-2.csv', delimiter=',')
d3 = genfromtxt('T1-3.csv', delimiter=',')
d4 = genfromtxt('T1-4.csv', delimiter=',')
```

```python
kmeans_1 = KMeans(n_clusters=3, random_state=0).fit(d1)
centroids = kmeans_1.cluster_centers_
cluster = kmeans_1.predict(d1)
plt.scatter(d1[:, 0], d1[:, 1], c=cluster)
plt.scatter(centroids[:,0],centroids[:,1],s=100,c="red")
plt.show()
```

Ciertamente lo esperado para esta situación sería que se hubiera formado un clúster vacío, dada la agrupación de los datos es esperable que se hubieran formado 2 clúster, cada uno en las agrupaciones visible, sin embargo, se produjeron 2 clúster en una de las agrupaciones, esto es posible si se considera que dentro de esta agrupación existan 2 subgrupos los cuales tengan una mejor relación entre ellos y es la razón por la cual k-means los agrupó de esa manera y así no dejar un clúster vació de manera que sea mas precisa la agrupación total.

```python
kmeans_2 = KMeans(n_clusters=3, random_state=0).fit(d2)
centroids = kmeans_2.cluster_centers_
cluster = kmeans_2.predict(d2)
plt.scatter(d2[:, 0], d2[:, 1], c=cluster)
plt.scatter(centroids[:,0],centroids[:,1],s=100,c="red")
plt.show()
```

Al igual que el caso anterior, era esperable encontrarnos con un clúster vacío, sin embargo, esto no ocurrió, en este caso, es posible deducir que los datos de los grupos grandes, (ambos círculos que se aprecian), posean datos que tenga una relación entre ellos, por lo tanto, k-means considera apropiado crear un clúster con un centroide justo en la intersección de ambos grupos y así generar estos 3 clúster de manera más precisa.

```python
kmeans_3 = KMeans(n_clusters=2, random_state=0).fit(d3)
centroids = kmeans_3.cluster_centers_
cluster = kmeans_3.predict(d3)
plt.scatter(d3[:, 0], d3[:, 1], c=cluster)
plt.scatter(centroids[:,0],centroids[:,1],s=100,c="red")
plt.show()
```

En este caso, si bien, era esperable que los centroides se ubicaran en ambas agrupaciones de datos (especies de elipses), sin embargo, por el funcionamiento de k-means, donde se busca la menor distancia al centroide para cada dato, considerar los centroides en ambas elipses generaría que los datos que se encuentran en los extremos quedaran muy alejados, por lo tanto, es más factible pensar que los centroides deban estar ubicados en un punto intermedio de ambas elipses, generando de esta manera los clúster obtenidos.

```python
kmeans_4 = KMeans(n_clusters=3, random_state=0).fit(d4)
centroids = kmeans_4.cluster_centers_
cluster = kmeans_4.predict(d4)
plt.scatter(d4[:, 0], d4[:, 1], c=cluster)
plt.scatter(centroids[:,0],centroids[:,1],s=100,c="red")
plt.show()
```

Por último, es parecido a lo ocurrido en el caso anterior, lo natural es pensar que cada centroide se ubicara en el centro de cada uno de los círculos, sin embargo, por cómo funciona k-means, los centroides se ven definidos por la cantidad de datos presentes, por lo tanto, el círculo central va a provocar que los centroides de los círculos más periféricos tiendan a acercarse más a él, por lo tanto, algunos datos del círculo central formarán parte de más de 1 clúster, que es lo que se observa en este caso.

```python
# Crear Figuras
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1,4, figsize=(12,3), dpi=120)

# Plot
ax1.scatter(d1[:,0], d1[:,1], s=1, marker=".")  
ax2.scatter(d2[:,0], d2[:,1], s=1) 
ax3.scatter(d3[:,0], d3[:,1], s=2) 
ax4.scatter(d4[:,0], d4[:,1], s=2)

ax1.set_title('(K=3)')
ax2.set_title('(K=3)')
ax3.set_title('(K=2)')
ax4.set_title('(K=3)')

ax1.set_xlim(-1,6)
ax1.set_ylim(-1,6)

ax2.set_xlim(-1,3)
ax2.set_ylim(-1,3)

plt.tight_layout()
plt.show()
```
